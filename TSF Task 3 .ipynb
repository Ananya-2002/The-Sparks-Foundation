## Corelation

# Shipping Mode

```python
df['Ship Mode'].value_counts()
```

```python
sns.countplot('Ship Mode', data = df,palette='Set2')
plt.xlabel('Ship Mode')
plt.ylabel('Count')
plt.title('Shipping Mode ',fontsize=20)
plt.show()
```

#### * Standard class is the most opted shipping mode

# Region

```python
df['Region'].value_counts()
```

```python
sns.countplot('Region',data=df,palette='Set2')
plt.xlabel('Region')
plt.ylabel('Buyers')
plt.title('Region wise buyers ',fontsize=20)
plt.show()
```

#### *West has the largest sales

# City

```python
df['City'].value_counts()
```

```python
plt.figure(figsize=(20,10))
sns.countplot(x=df['City'],order=(df['City'].value_counts().head(50)).index)
plt.xticks(rotation=80)
plt.xlabel('City')
plt.ylabel('Buyers')
plt.title('City wise buyers ',fontsize=20)
plt.show()
```

#### *New York has the highest number of buyers

# State

```python
df['State'].value_counts()
```

```python
plt.figure(figsize=(20,10))
sns.countplot(x=df['State'],order=(df['State'].value_counts().head(50)).index)
plt.xticks(rotation=80)
plt.xlabel('State')
plt.ylabel('Buyers')
plt.title('State wise buyers ',fontsize=20)
plt.show()
```

#### *California has the highest number of buyers

# Category

```python
df['Category'].value_counts()
```

```python
sns.barplot(x='Category',y='Sales',data=df,palette='Set2')
plt.ylabel('Sales')
plt.title('Category wise sales')
plt.show()

```
